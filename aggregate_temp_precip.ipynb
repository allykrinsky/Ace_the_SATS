```python
#reverse std
avg_20_precip = []
avg_20_temp = []
for val in df.iterrows():

    #print(val[1][4])
    temp_std = val[1][4]
    temp_ma = val[1][3]

    precip_std = val[1][7]
    precip_ma = val[1][6]

    avg_20_temp.append(-(np.sqrt((temp_std ** 2 ) * 5) - temp_ma))
    avg_20_precip.append(-(np.sqrt((precip_std ** 2 ) * 5) - precip_ma))

```

```python
df['temp_avg_20'] = avg_20_temp
df['precip_avg_20'] = avg_20_precip
```
