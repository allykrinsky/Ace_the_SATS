```python
import pandas as pd
climate = pd.read_csv('seasonal_disparities_20th_century.csv')
climate = climate[['FIPSCode','YEAR', 'Season', 'TEMPERATURE_MA', 'TEMPERATURE_DEVIATION_STDDEV', 'PRECIPITATION_MA', 'PRECIPITATION_DEVIATION_STDDEV']]
climate.head()
```

```python
import numpy as np
temp_chng = []
precip_chng = []
for val in climate.iterrows():

    temp_std = val[1][4]
    temp_ma = val[1][3]

    precip_std = val[1][6]
    precip_ma = val[1][5]


    avg_20_temp= -(np.sqrt((temp_std ** 2 ) * 5) - temp_ma)
    avg_20_precip = -(np.sqrt((precip_std ** 2 ) * 5) - precip_ma)

    temp_chng.append((temp_ma - avg_20_temp) / avg_20_temp * 100)
    precip_chng.append((precip_ma - avg_20_precip) / avg_20_precip * 100)
    
```

```python
climate['temp_avg_20'] = temp_chng
climate['precip_avg_20'] = precip_chng
```

```python
climate = climate[['FIPSCode', 'YEAR', 'Season', 'temp_avg_20', 'precip_avg_20']]
```
